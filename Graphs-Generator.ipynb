```python
factor = (1.23984E-4) #accounting for conversion between ev to cm^-1
k = 8.6173303E-5 #boltzmann constant in eV/K
#k =0.6950389 # in cm^-1. Change made on 8th of April
T = 300 #Temperature in K
hbar = 0.658211951 # in eV-fs

Gammak =  0.00062/hbar #9.4196*10**(-4) #rad/fs

#Putting Peaks together
def Lorentzian(ww):
    result = 0
    for i in range(62):
        result += (2*hbar/pi)*(Sk[i]*Wk[i]**3*ww*Gammak)/((Wk[i]**2-ww**2)**2+(ww**2*Gammak**2))
    return result

#Frequencies and factors for 62 Lorentzian Peaks from "Lorentzian" paper
Sk = np.array([0.011, 0.011, 0.009, 0.009, 0.010, 0.011, 0.011, 0.012, 0.003, 0.008,
         0.008, 0.003, 0.006,  0.002, 0.002, 0.002,  0.001, 0.002, 0.004,  0.007, 
         0.004, 0.004, 0.003,  0.006, 0.004, 0.003, 0.007, 0.010, 0.005, 0.004, 
         0.009, 0.018,  0.007, 0.006, 0.007, 0.003, 0.004, 0.001,  0.001,  0.002, 
         0.002,  0.003, 0.001, 0.002, 0.002, 0.001, 0.001, 0.003, 0.003, 0.009, 0.007,
         0.010, 0.003, 0.005, 0.002, 0.004, 0.007, 0.002, 0.004, 0.002, 0.003, 0.003])
#array with all wk
Wk = np.array([46, 68, 117, 167, 180, 191, 202, 243, 263, 284, 291, 327, 366, 385, 404, 423, 440, 
         481, 541, 568, 582, 597, 630, 638, 665, 684, 713, 726, 731, 750, 761, 770, 795, 821,
         856, 891, 900, 924, 929, 946, 966, 984, 1004, 1037, 1058, 1094, 1104, 1123, 1130, 1162,
         1175, 1181, 1201, 1220, 1283, 1292, 1348, 1367, 1386, 1431, 1503, 1545])*1.23984*10**(-4)/hbar
#with this factors we go from cm-1 -> ev -> rad/fs
```

```python
ww = np.linspace(0,2,100000)
Lorentz = Lorentzian(ww)

plt.figure()
plt.plot(ww*hbar,Lorentz,'r',label='Theoretical') #change by Lorentz/factor if we want cm-1
plt.xlabel('$\hbar\omega$ [eV]')
plt.ylabel('$J(\omega)$ [$eV$]')
#ylim(-0.1,20*10**(-0))
plt.xlim(0.005,0.25) 

plt.legend()
plt.show()
```
